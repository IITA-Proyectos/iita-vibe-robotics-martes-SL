# Control de flujo

Todo programa informático está formado por *instrucciones* que se ejecutan en forma secuencial de arriba hacia abajo, como si estuviésemos leyendo un libro. Este orden constituye el llamado **flujo** del programa. 

Es posible modificar este flujo secuencial para que tome *bifurcaciones* o *repita* ciertas instrucciones. Las sentencias que nos permiten hacer estas modificaciones se engloban en el **control de flujo**.

## Comentarios

Los comentarios son anotaciones que podemos incluir en nuestro programa y que nos van a permitir aclarar ciertos aspectos del código. Estas indicaciones son ignoradas por el intérprete de Python.

Los comentarios se incluyen usando el símbolo almohadilla **#** y comprenden hasta el final de la línea.

Veamos un comentario en bloque:

```python
# Edad del Universo expresando en días
edad_del_universo = 13800 * (10 ** 6) * 365
```

Los comentarios también pueden aparecer en la misma línea de código, aunque la guía de Python no aconseja usarlos demasiado.

Comentario en línea:

```python
stock = 0 # Stock actual
```

Reglas para **escribir buenos comentarios**:

1. Los comentarios no deberían duplicar el código.
2. Los buenos comentarios no van a arreglar un código con poca claridad.
3. Si no puedes escribir un comentario claro, puede haber un problema en el código.
4. Los comentarios deberían evitar la confusión, no crearla.
5. Usa comentarios para explicar código no idiomático.
6. Proporciona enlaces a la fuente original del código copiado.
7. Incluye enlaces a referencias externas que sean de ayuda.
8. Añade comentarios cuando arregles errores.
9. Usa comentarios para destacar implementaciones incompletas.

## La sentencia if:

La sentencia condicional en Python (al igual que en muchos otros lenguajes de programación) es **if**. En su escritura debemos añadir una **expresión de comparación** terminando con **dos puntos al final** de la línea. Veamos un ejemplo:

```python
temperatura = 40
if temperatura > 35:
    print('Aviso por alta temperatura')
```

Notemos que en Python no es necesario incluir paréntesis ( y ) al escribir condiciones. Hay veces que es recomendable por claridad o por establecer prioridades.

En el caso anterior se puede ver claramente que la condición se cumple y por tanto se ejecuta la instrucción que tenemos dentro del cuerpo de la condición. Pero podría no ser así. 

Para controlar ese caso existe la sentencia **else**. Veamos el mismo ejemplo anterior pero añadiendo esta variante:

```python
temperatura = 20
if temperatura > 35:
    print('Aviso por alta temperatura')
else:
    print('Parámetros normales')
```

Podríamos tener incluso condiciones dentro de condiciones, lo que se viene a llamar técnicamente **condiciones anidadas**. 

El **anidamiento** (o nesting) hace referencia a incorporar sentencias unas dentro de otras mediante la inclusión de diversos niveles de profundidad (indentación).

Veamos un ejemplo ampliando el caso anterior:

```python
temperatura = 28
if temperatura < 20:
    if temperatura < 10:
        print('Nivel azul')
    else:
        print('Nivel verde')
else:
    if temperatura < 30:
        print('Nivel naranja')
    else:
        print('Nivel rojo')
```

Python nos ofrece una mejora en la escritura de condiciones anidadas cuando aparecen consecutivamente un **else** y un **if**. Podemos sustituirlos por la sentencia **elif**:

![](docs/19elif.png)

Apliquemos esta mejora al código anterior:

```python
temperatura = 28
if temperatura < 20:
    if temperatura < 10:
        print('Nivel azul')
    else:
        print('Nivel verde')
elif temperatura < 30:
    print('Nivel naranja')
else:
    print('Nivel rojo')
```

## Asignaciones condicionales

Supongamos que queremos asignar un nivel de riesgo de incendio en función de la temperatura. En su **versión clásica** escribiríamos:

```python
temperatura = 35
if temperatura < 30:
    RIESGO_INCENDIO = 'BAJO'
else:
    RIESGO_INCENDIO = 'ALTO'
RIESGO_INCENDIO
```

Sin embargo, esto lo podríamos abreviar con una **asignación condicional de una única línea**:

```python
RIESGO_INCENDIO = 'BAJO' if temperatura < 30 else 'ALTO'
RIESGO_INCENDIO
```

## Cortocircuito lógico

Es interesante que también comprendamos que **las expresiones lógicas no se evalúan por completo si se dan una serie de circunstancias**. Aquí es cuando nos referimos al concepto de **cortocircuito**, que es la forma en la que vamosa  denominar estos escenarios.

Supongamos un ejemplo en el que utilizamos un **celular** que mide la batería por la variable **power** de 0% a 100% y la señal 4G por la variable signal_4g de 0 a 100% también.

Ahora en el caso de que **necesitemos enviar un mensaje por Telegram**, se debe cumplir que el teléfono tenga al menos un 25% de batería y al menos un 10% de señal:

```python
power = 10
signal_4g = 60
```

```python
power > 25 and signal_4g > 10
```

Si analizamos la expresión, nos encontraremos con un **and** y la primera condición **power > 25 no se cumple**, entonces se produce un cortocircuito y **no se sigue evaluando el resto de la expresión** porque ya se sabe que va a dar False.

![](docs/21cortoand.png)

Veamos otro ejemplo. Para poder hacer una **videollamada** necesitamos tener al menos un 40% de batería o al menos un 30% de cobertura:

```python
power = 50
signal_4g = 20
```

```python
power > 40 or signal_4g > 30
```

![](docs/22cortoor.png)

En este caso estamos en un **or** y la primera condición **power > 40** se cumple, entonces no se va a seguir evaluando el resto de la expresión porque **ya se sabe que va a dar True**.

> *Obviamente si no se produce ningún cortocircuito en la evaluación de la expresión, se seguirá comprobando todas las condiciones posteriores hasta llegar al final de la misma.*

## Booleanos en condiciones

Cuando queremos preguntar por la **veracidad** de una determinada variable booleana en una condición, la primera idea que nos encontramos es la siguiente:

```python
hace_frio = True
```

```python
if hace_frio == True:
    print('Hay que abrigarse')
else:
    print('Con una remera estoy bien')
```

También podemos **simplificar** esta condición para que quede así:

```python
if hace_frio:
    print('Hay que abrigarse')
else:
    print('Con una remera estoy bien')
```

Hicimos esta comparación para un valor booleano verdadero (True). En el caso de que la comparación fuera para un valor falso lo haríamos así:

```python
hace_frio = False
```

```python
if not hace_frio:
    print('Con una remera estoy bien')
else:
    print('Hay que abrigarse')
```

Si lo analizamos nos daremos cuenta que estamos reproduciendo bastante bien y entendible nuestro *lenguaje natural*:

- Si hace frío, nos abrigamos.
- Si no hace frío, con una remera estaremos bien.

## Valor nulo

**None** es un valor especial en Python que almacena el **valor nulo**. Veamos con ejemplos como se comporta este en condiciones de *veracidad*:

```python
valor = None
if valor:
    print('la variable valor debe tener un valor útil')
else:
    # valor podría contener None, False (u otro)
    print('la variable valor parece tener un valor nulo')
```

Para distinguir **None** de los valores propiamente booleanos, se recomienda usar el operador *is*. Veamos un ejemplo para averiguar si un valor es **nulo**:

```python
if valor is None:
    print('Claramente es nulo')
else:
    # Quizá valor es un True, False u otro dato
    print('valor tiene algo útil')
```

De igual forma, podemos usar esta construcción para el caso contrario. La forma pitónica de preguntar si algo **no es nulo** es la siguiente:

```python
valor = 99
```

```python
if valor is not None:
    print(f'{valor=}')
```

Si nos preguntamos porqué utilizamos *is* en vez del operador == al comprobar si un valor es nulo, veamos porqué:

```python
valor = None
```

```python
valor is None
```

```python
valor == None
```

La respuesta es que el operador is comprueba únicamente si los identificadores (posiciones en memoria) de dos objetos son iguales, mientras que la comparación == puede englobar muchas otras acciones. De este hecho se deriva que su ejecución sea mucho más rápida y que
se eviten **falsos positivos**.

Cuando ejecutamos un programa Python existe una serie de objetos precargados en memoria. Uno de ellos es **None**:

```python
id(None)
```

Cualquier variable que igualemos al valor nulo, únicamente **será una referencia al mismo objeto None** en memoria:

```python
valor = None
id(valor)
```

Por lo tanto, ver si un objeto es **None** es simplemente comprobar que su identificador coincida con el de None, que es exactamente el cometido de la función **is()**:

```python
id(valor) == id(None)
```

```python
valor is None
```

Como dato interesante, Python carga inicialmente en memoria objetos como True o False, pero también los números enteros que van desde el -5 hasta el 256. Se entiende que tiene que ver con optimizaciones a nivel de rendimiento.

Lo primero que debemos entender de cara a comprobar la **veracidad** son los valores que **evalúan a falso** o **evalúan a verdadero**.

Veamos las únicas cosas que son evaluadas a False en Python:

```python
bool(False)
```

```python
bool(None)
```

```python
bool(0)
```

```python
bool(0.0)
```

```python
bool('') # Cadena vacía
```

```python
bool([]) # Lista vacía
```

```python
bool(()) #Tupla vacía
```

```python
bool({}) # Diccionario vacío
```

```python
bool(set()) # Conjunto vacío
```
